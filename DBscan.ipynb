```python
%matplotlib inline
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
```

```python
df = pd.read_csv(r"telemetry.csv")
```

```python
df_9 = df.loc[df['machineID']==40]
df_9
```

```python
error = pd.read_csv(r"errors.csv")
```

```python
error_9 = error.loc[error['machineID']==40]
```

```python
failure = pd.read_csv(r'failures.csv')
```

```python
failure_9 = failure.loc[failure['machineID']==40]
```

```python
error_9
```

```python
merged_9 = df_9.merge(error_9,how='outer', left_on='datetime',right_on='datetime')
```

```python
merged_9.head(120)
```

```python
merged_9['errorID'].fillna("None",inplace=True)
```

```python
merged_9
```

```python
merged_9.drop('machineID_y',axis=1,inplace=True)
merged_9
```

```python
merged_9.replace(to_replace=['None','error1','error2','error3','error4','error5'],value=[0,1,2,3,4,5],inplace=True)
```

```python
merged_9
```

```python
from sklearn.cluster import DBSCAN
import sklearn.utils
from sklearn.preprocessing import StandardScaler
sklearn.utils.check_random_state(1000)
#Clus_dataSet = merged[['vibration','rotate','errorID','volt']]
Clus_dataSet = merged_9[['volt','rotate','vibration','errorID']]
#Clus_dataSet = merged_9[['volt','errorID']]
#Clus_dataSet = np.nan_to_num(Clus_dataSet)
Clus_dataSet = StandardScaler().fit_transform(Clus_dataSet)

# Compute DBSCAN
db = DBSCAN(eps=4, min_samples=3).fit(Clus_dataSet)
core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
core_samples_mask[db.core_sample_indices_] = True
labels = db.labels_
merged_9["Clus_Db"]=labels

realClusterNum=len(set(labels)) - (1 if -1 in labels else 0)
clusterNum = len(set(labels)) 


# A sample of clusters
merged_9[['volt','rotate','pressure','vibration','errorID','Clus_Db']].head(120)
```

```python
y_pred = db.fit_predict(Clus_dataSet)
y_pred
```

```python
plt.scatter(Clus_dataSet[:,0], Clus_dataSet[:,1],c=y_pred,marker='x')
plt.title("DBSCAN")
```

```python
merged_9.to_csv("merged-9.csv")
```

```python
merged_9 = merged_9.merge(failure_9,how='outer', left_on='datetime',right_on='datetime')
merged_9
```

```python
merged_9.drop('machineID',axis=1,inplace=True)
merged_9
```

```python
merged_9['failure'].fillna("None",inplace=True)
```

```python
merged_9.replace(to_replace=['None','comp1','comp2','comp3','comp4'],value=[0,1,2,3,4],inplace=True)
```

```python
from sklearn.cluster import DBSCAN
import sklearn.utils
from sklearn.preprocessing import StandardScaler
sklearn.utils.check_random_state(1000)
#Clus_dataSet = merged[['vibration','rotate','errorID','volt']]
Clus_dataSet_f = merged_9[['volt','rotate','vibration','errorID','failure']]
#Clus_dataSet_f = merged_9[['volt','failure']]
#Clus_dataSet = np.nan_to_num(Clus_dataSet)
Clus_dataSet_f = StandardScaler().fit_transform(Clus_dataSet_f)

# Compute DBSCAN
db = DBSCAN(eps=3.8, min_samples=3).fit(Clus_dataSet_f)
core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
core_samples_mask[db.core_sample_indices_] = True
labels = db.labels_
merged_9["Clus_Db_f"]=labels

realClusterNum=len(set(labels)) - (1 if -1 in labels else 0)
clusterNum = len(set(labels)) 


# A sample of clusters
merged_9[['volt','rotate','pressure','vibration','errorID','Clus_Db','failure','Clus_Db_f']].head(120)
```

```python
y_pred_f = db.fit_predict(Clus_dataSet_f)
y_pred_f
```

```python
plt.scatter(Clus_dataSet_f[:,0], Clus_dataSet_f[:,1],c=y_pred_f,marker='x')
plt.title("DBSCAN")
```

```python
merged_9.to_csv("merged-9.csv")
```
